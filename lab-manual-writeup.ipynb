## Testing Participants

### From start to finish, testing participants should proceed with the following order:

1) Show up at least 15 minutes before the scheduled testing time to set-up (grab consent forms, prepare testing room, pre-load code on computer, etc.)
2) Greet the participant and introduce yourself to them when they arrive
3) Have the participant read and sign the consent form
4) Bring the participant into testing room and go through pre-experimental instructions (see more details in 'Experiment Instructions' document)
5) Start the program code once the participant is ready to begin testing
6) Walk the participant through baseline trials to ensure proper execution and complete understanding of the experimental task (see more details in 'Experiment Instructions' document)
7) Monitor the participant's behavior throughout the experiment; correct improper behavior if needed
    - If anything notable (that could potentially compromise data quality) occurs, be sure to record it in the 'Lab Notebook' document
8) Once the participant completes the experiment, hit the *Esc* key to automatically close the program and save the data
9) Ask the participant end of experiment questions; record responses in the 'Lab Notebook' document
10) Pay the participant and have them sign the payment confirmation form
11) Thank the participant for their time and wish them a good day before they leave!

**Consent Forms:**
